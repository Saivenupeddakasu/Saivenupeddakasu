```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!ls
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
df.shape
```

```python
df.isnull()
```

```python
df.isnull().sum() #return null values
```

```python
df.dropna(inplace=True)
# fix some values  inplace of in if data set  contains more null values
# if less null values easily deleted the null values
```

```python
df.shape
```

```python
df.columns # all columns
```

```python
#change type
df['Amount']=df['Amount'].astype('int')
```

```python
df.rename(columns={'Marital_Status':'Shaadi'}) # for rename
```

```python
#Exploratory Data anyalsis
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a bar chart for gender vs total amount
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
ax=sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
#age group Vs gender
```

```python
ax=sns.countplot(data=df, x='State', hue='Zone')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='State', y='Amount', data=sales_age)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
sales_age = df.groupby(['Age Group', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', hue='Gender', data=sales_age)
#
```

```python
ax = sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax = sns.countplot(data=df, x='State')
sns.set(rc={'figure.figsize':(6,10)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax=sns.countplot(data=df, x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x='Product_Category', y='Amount')
```
